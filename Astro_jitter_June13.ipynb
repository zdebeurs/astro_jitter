```python
import os

import matplotlib.pyplot as plt

import astropy.time
import astropy.units as u
from astropy.coordinates import SkyCoord
from astropy.visualization import ImageNormalize, SqrtStretch

import sunpy.map
from sunpy.net import Fido
from sunpy.net import attrs as a
```

```python
t0 = astropy.time.Time('2012-09-24T14:56:03', scale='utc', format='isot')
query = Fido.search(
    a.Instrument.aia,
    a.Physobs.intensity,
    a.Time(t0, t0 + 13*u.s),
)
files = Fido.fetch(query)
amap = sunpy.map.Map(files)
```

```python
amap = sunpy.map.Map(files[0])
```

```python
for i in range(0, 5):
    
    amap = sunpy.map.Map(files[i])
    cutouts = amap.submap(
        SkyCoord(-500*u.arcsec, -275*u.arcsec, frame=amap.coordinate_frame),
        top_right=SkyCoord(150*u.arcsec, 375*u.arcsec, frame=amap.coordinate_frame),
    )
    plt.figure()
    cutouts.plot()

    plt.show()
```
